### **Correlations in each year**

```python
for year, df in dfs.items():
    print(f"\nCorrelation Analysis {year}")
    df_temp = df.copy()

    # Select only numeric columns
    df_num = df_temp.select_dtypes(include='number')

    target_col = None
    for col in df_num.columns:
        if col.lower() in ['score', 'happiness score', 'happiness_score']:
            target_col = col
            break
    if not target_col:
        print("No 'Score' column found.")
        continue

    # Correlation matrix
    corr_matrix = df_num.corr()

    # Sort correlations relative to the Score
    corr_with_score = corr_matrix[target_col].sort_values(ascending=False)
    print("\nTop 10 correlations with Score:")
    print(corr_with_score.head(10))
    print("\nBottom 10 correlations with Score:")
    print(corr_with_score.tail(10))

    # Heatmap of main feature correlations
    main_features = [c for c in df_num.columns if not c.startswith("Country_")]
    plt.figure(figsize=(10, 6))
    sns.heatmap(
        df_num[main_features].corr()[[target_col]].sort_values(by=target_col, ascending=False),
        annot=True, cmap='coolwarm', fmt='.2f'
    )
    plt.title(f'Correlation Matrix with {target_col} - {year}')
    plt.show()
```

```python
def non_repeated_countries(*dfs):
    
    sets = [set(df['Country']) for df in dfs]
    common = set.intersection(*sets)
    all_countries = set.union(*sets)
    non_repeated = all_countries - common

    print(f"Total unique countries across all years: {len(all_countries)}")
    print(f"Countries present in all years: {len(common)}")
    print(f"Countries missing in at least one year: {len(non_repeated)}")
    
    return non_repeated
```

```python
different_countries = non_repeated_countries(df_2015, df_2016, df_2017, df_2018, df_2019)
print("Countries not present in all five years:\n", different_countries)
```

```python
def normalize_country(name):
    name = name.strip().lower()
    replacements = {
        'hong kong s.a.r., china': 'hong kong',
        'taiwan province of china': 'taiwan',
        'somaliland region': 'somaliland',
        'north macedonia': 'macedonia',
        'swaziland': 'eswatini',
        'trinidad and tobago': 'trinidad & tobago',
        'north cyprus': 'northern cyprus'
    }
    return replacements.get(name, name)
```

### **Merge**

```python
final_columns = [
    'Country', 'Freedom', 'GDP per Capita', 'Generosity',
    'Healthy life expectancy', 'Perceptions of corruption',
    'Score', 'Social support', 'Year'
]

# Concatenate all DataFrames 
df_all = pd.concat(
    [df_2015[final_columns],
     df_2016[final_columns],
     df_2017[final_columns],
     df_2018[final_columns],
     df_2019[final_columns]],
    ignore_index=True
)

# Apply country name normalization
df_all['Country'] = df_all['Country'].apply(normalize_country)

print(df_all.info())
print(df_all.head())
```

### **General validity reviews**

```python
for col in numeric_cols:
    Q1 = df_all[col].quantile(0.25)
    Q3 = df_all[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_all[(df_all[col] < lower_bound) | (df_all[col] > upper_bound)]
    print(f"{col}: {len(outliers)} possible outliers")
```

```python
countries = df_all['Country'].unique()

def validate_country(country_name):
    try:
        response = requests.get(f"https://restcountries.com/v3.1/name/{country_name}?fullText=true")
        if response.status_code == 200:
            return True
        else:
            return False
    except Exception as e:
        print(f"Error checking {country_name}: {e}")
        return False

result = pd.DataFrame({
    "Country": countries,
    "Is_valid": [validate_country(c) for c in countries]
})

invalid_countries = result[result["Is_valid"] == False]
print("Countries that do NOT exist according to the API:\n", invalid_countries)
```

```python
invalid_countries = [
    'Trinidad & Tobago',
    'Northern Cyprus',
    'Somaliland',
    'Macedonia',
    'Palestinian Territories',
    'Congo (Kinshasa)',
    'Congo (Brazzaville)'
]

print("--- Presence by year ---")
for country in invalid_countries:
    years = sorted(df_all.loc[df_all['Country'].str.lower() == country.lower(), 'Year'].unique())
    print(f"{country:<25} -> Years present: {years}")
```
